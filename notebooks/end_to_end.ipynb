```python
# Install once in your env:
# uv pip install cosmograph pandas  # or: pip install cosmograph pandas

import pandas as pd
from cosmograph import cosmo  # official widget helper

points = pd.DataFrame({
    "id":    ["a", "b", "c"],
    "label": ["Node A", "Node B", "Node C"],
    "group": ["X", "X", "Y"],
})
links = pd.DataFrame({
    "source": ["a", "b", "c"],
    "target": ["b", "c", "a"],
    "weight": [1.0, 1.0, 1.0],
})

w = cosmo(
    points=points,
    links=links,
    point_id_by="id",
    link_source_by="source",
    link_target_by="target",
    point_label_by="label",     # <- turn labels on
    point_color_by="group",
    point_include_columns=["group"],
    link_include_columns=["weight"],
)
w  # renders inline
```
